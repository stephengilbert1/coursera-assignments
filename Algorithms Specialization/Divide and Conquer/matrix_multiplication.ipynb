Problem: Matrix Multiplication
Input: Two n ⇥ n integer matrices, X and Y.
6
Output: The matrix product X· Y.

Straightforward MAtrix Multiplication

```python
def mat_mul(X,Y):
    n = len(X)
    Z = [[0] * n for _ in range(n)]
    for i in range(n):
        for j in range(n):
            Z[i][j] = 0
            for k in range(n):
                Z[i][j] = Z[i][j] + X[i][k]*Y[k][j]
    return Z

X1 = [[1, 2], [3, 4]]
Y1 = [[5, 6], [7, 8]]
print(mat_mul(X1,Y1))

X2 = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
Y2 = [[9, 8, 7], [6, 5, 4], [3, 2, 1]]
print(mat_mul(X2,Y2))

X3 = [[2, 0, 0], [0, 3, 0], [0, 0, 4]]  # Diagonal
Y3 = [[1, 0, 0], [0, 1, 0], [0, 0, 1]]  # Identity
print(mat_mul(X3,Y3))
```

Recursive algorithm for matrix multiplication

```python
def MatAdd(X,Y):
    n=len(X)
    
    Z = [[0] * n for _ in range(n)]
    for i in range(n):
        for j in range(n):
            Z[i][j] = X[i][j] + Y[i][j]
    return Z

def RecMatMult(X,Y):
    n = len(X)
    
    if n == 1:
        Z = [[X[0][0] * Y[0][0]]]
        return Z
    
    else:
        m = n//2

        A = [[0] * m for _ in range(m)]
        B = [[0] * m for _ in range(m)]
        C = [[0] * m for _ in range(m)]
        D = [[0] * m for _ in range(m)]
        E = [[0] * m for _ in range(m)]
        F = [[0] * m for _ in range(m)]
        G = [[0] * m for _ in range(m)]
        H = [[0] * m for _ in range(m)]

        for i in range(m):
            for j in range(m):
                A[i][j] = X[i][j]                    # Top-left
                B[i][j] = X[i+m][j]                  # Bottom-left
                C[i][j] = X[i][j+m]                  # Top-right
                D[i][j] = X[i+m][j+m]                # Bottom-right
                E[i][j] = Y[i][j]
                F[i][j] = Y[i+m][j]
                G[i][j] = Y[i][j+m]
                H[i][j] = Y[i+m][j+m]

        AE = RecMatMult(A,E)
        BG = RecMatMult(B,G)
        CE = RecMatMult(C,E)
        DG = RecMatMult(D,G)
        AF = RecMatMult(A,F)
        BH = RecMatMult(B,H)
        CF = RecMatMult(C,F)
        DH = RecMatMult(D,H)

        z11 = MatAdd(AE,BG)
        z12 = MatAdd(AF,BH)
        z21 = MatAdd(CE,DG)
        z22 = MatAdd(CF,DH)

        Z = [[0] * n for _ in range(n)]

        for i in range(m):
            for j in range(m):
                Z[i][j] = z11[i][j]
                Z[i][j+m] = z12[i][j]
                Z[i+m][j] = z21[i][j]
                Z[i+m][j+m] = z22[i][j]

        return Z


print(RecMatMult(X1,Y1))
print(RecMatMult(X3,Y3))
```
