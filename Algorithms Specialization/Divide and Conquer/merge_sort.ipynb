Merge Sort (for practice)

```python
def merge(left,right):
    n = len(left) + len(right)
    i = 0
    j = 0
    B = []
    while i < len(left) and j < len(right):
        if left[i] < right[j]:
            B.append(left[i])
            i+=1
        else:
            B.append(right[j])
            j+=1
    B.extend(left[i:])
    B.extend(right[j:])
    return B

print(merge([1,3,9], [2,4,6]))
```

```python
def merge_sort(arr):
        if len(arr) <= 1:
            return arr
        mid = len(arr)//2
        left = merge_sort(arr[:mid])
        right = merge_sort(arr[mid:])
        return merge(left,right)

print(merge_sort([9,1,2,7,3,4,3,5,6]))
```
