Grade-schooo method with recursive integer multiplication

```python
def split(num):
    s = str(num)
    mid = len(s) //2
    #print (mid)
    high = int(s[:mid])if mid > 0 else 0
    low = int(s[mid:])
    return high, low
    

print(split(175))
```

```python
def multiply(x,y):
    #print(f"multiply({x}, {y})")
    if x < 10 and y < 10:
        return x * y

    x1,x0 = split(x)
    y1,y0 = split(y)
    #print(f"  splits: x={x1},{x0}, y={y1},{y0}")

    p0 = multiply(x0,y0)
    p1 = multiply(x1,y0) + multiply(x0,y1)
    p2 = multiply(x1,y1)

    m = len(str(x)) // 2
    #print(f"m = {m}")
    #print(f"p2={p2}, p1={p1}, p0={p0}")
    return  p2*10**(2*m)+p1*10**m+p0


#print(multiply(12,34))
#print(12*34)
print(f"Algorithm output: {multiply(12345678,23456789)}")
print(f"Check:{12345678*23456789}")
print(f"diff = {multiply(12345678,23456789) - (12345678*23456789)}")
```

Karatsuba Algorithm next

```python
def split_at(num, m):
    s = str(num).zfill(2*m)

    high = int(s[:m])
    low = int(s[m:]) 
    return high, low
    
```

```python
def karatsuba(x, y):
    if x<10 and y<10:
        return x*y
    
    m = max(len(str(x)),len(str(y)))
    m2 = (m+1)//2

    x1,x0 = split_at(x,m2)
    y1,y0 = split_at(y,m2)
    #print(f"Splits: x={x1},{x0}, y={y1},{y0}")

    z0 = karatsuba(x0,y0)
    z1 = karatsuba(x0+x1, y0+y1)
    z2 = karatsuba(x1,y1)

    return (z2*10**(2*m2))+((z1-z2-z0)*10**m2)+z0

print(f"Karatsuba Algorithm Output: {karatsuba(num1,num2)}")
print(f"Check:{num1*num2}")
print(f"Diff: {karatsuba(num1,num2) - (num1*num2)}")
```
