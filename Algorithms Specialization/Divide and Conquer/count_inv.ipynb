Problem: Counting Inversions

Input: An array A of distinct integers.
Output: The number of inversions of Aâ€”the number of
pairs (i, j) of array indices with i < j and A[i] > A[j].

```python
def mergeCountSplitInv(sorted_left, sorted_right):
    i = 0
    j = 0
    splitInv = 0
    b = []
    n = len(sorted_left) + len(sorted_right)
    for k in range(n):
        if i < len(sorted_left) and (j>= len(sorted_right) or sorted_left[i]<sorted_right[j]):
            b.append(sorted_left[i])
            i+=1
        else:
            b.append(sorted_right[j])
            j+=1
            splitInv+= (len(sorted_left) -i)
    return b, splitInv
```

```python
def count_inv(arr):
    n = len(arr)
    if n<=1:
        return arr, 0
    else:
        midPoint = n//2
        left, leftInv = count_inv(arr[:midPoint])
        right, rightInv = count_inv(arr[midPoint:])
        merged, splitInv = mergeCountSplitInv(left, right)
        return merged, leftInv+rightInv+splitInv
    
count_inv([6,5,4,3,2,1])
```
